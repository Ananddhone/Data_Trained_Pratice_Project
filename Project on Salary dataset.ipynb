# Salary Dataset
Salary dataset related to salary of the employee from various team. Where in the dataset we can get the details of the employee. so let's analysis the dataest of employee and further impact.
#link for dataset:https://github.com/dsrscientist/DSData/blob/bab2ac243722e052c220244e80c06433c902c123/MNC_Salaries.csv#L16

```python
import pandas as pd
df_salary=pd.read_csv('salary.csv')
df_salary
```

```python
#let's get a quick overview of the dataset
df_salary.info()
```

```python
#Let's calculating some statistical data like percentile, mean and std of salary dataset. 
df_salary.describe()
```

```python
# for getting accurate result of any dataset need to check wheter any null value present in dataset or not.
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df_salary.isnull(),annot=True)
plt.show()
```

```python
#Checking the null values without using visualisation.
df_salary.isnull().sum()
```

```python
#let's check the skewness of the dataset 
df_salary.skew()
#skewness is 0, the data are perfectly symmetrical, although it is quite unlikely for real-world data. 
#As a general rule of thumb: If skewness is less than -1 or greater than 1, the distribution is highly skewed.
```

```python
#let's see that how salary change with age of employee.
sns.lineplot(x=df_salary["Age"],y=df_salary["Salary"],data=df_salary)
plt.show()
#We can see that as the age of employee get up salary also get up but at one point of time in between 35 to 40 salary getting down.
```

```python
#let's see that how salary change with Position of employee.
sns.lineplot(x=df_salary["Position"],y=df_salary["Salary"],data=df_salary)
plt.show()
```

```python
#Average salary as per the age
sns.barplot(x=df_salary["Age"],y=df_salary["Salary"],data=df_salary)
plt.show()
```

```python
#Average salary as per the age Position
sns.barplot(x=df_salary["Position"],y=df_salary["Salary"],data=df_salary)
plt.show()
```

```python
#Average salary as per the age Team
sns.barplot(x=df_salary["Salary"],y=df_salary["Team"],data=df_salary)
plt.show()
```
